# 坐标轴

```python
%matplotlib inline
import matplotlib as mpl 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import random

mpl.pyplot.rcParams["font.sans-serif"] = "SimHei" # 解决中文乱码
mpl.pyplot.rcParams["axes.unicode_minus"] = False # 解决负号显示问题

df = pd.DataFrame({"Month":[1,2,3,4,5,6,7,8,9], 
                   "Registration":[866,2335,5170,6482,6120,1605,3813,4428,4631],
                   "Activation":[433,1167,2855,3241,3060,802,1902,2214,2315]})
df


# plt.xlabel("月份")
# plt.ylabel("注册量")
```

# 3. tick 刻度
* `ticks`: 刻度位置
* `labels`: 刻度处显示的内容

```python
# plt.xtick(ticks,labels)
# plt.ytick(ticks,labels)
plt.xticks([1,2,3,4,5,6,7,8,9],["Jan","Feb","Mar","Apr","May","Jun","July","Aug","Sept"])
plt.xlabel("Month", labelpad = 10) # tick label和axes的距离
plt.ylabel("Activation", labelpad = 10)
```

```python
plt.xticks(np.arange(1,11),["Jan","Feb","Mar","Apr","May","Jun","July","Aug","Sept","Oct"])
plt.yticks(np.arange(0,7000,1000),["1000 people","2000 people","3000 people","4000 people","5000 people","6000 people","7000 people"])
plt.xlabel("Month", labelpad = 10) # tick label和axes的距离
plt.ylabel("Activation", labelpad = 10)
plt.plot(df["Month"],df["Activation"])
```

### 隐藏tick

```python
plt.xticks([])
plt.yticks([])
plt.xlabel("Month", labelpad = 10) # tick label和axes的距离
plt.ylabel("Activation", labelpad = 10)
plt.plot(df["Month"],df["Activation"])
```

 ## 综合设置
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tick_params.html

`matplotlib.pyplot.tick_params(axis='both', **kwargs)`
* *args称之为Non-keyword Variable Arguments，无关键字参数：
    * ***args - 不定长参数：list set tuple dictionary**

* **kwargs称之为keyword Variable Arguments，有关键字参数；
    * dictionary 传入

```python
# which 设置 主or次 tick线
# direction tick 朝向
# bottom 下刻度线是否显示 没有在坐标轴上标出竖线
plt.tick_params(axis = "both", which = "both", direction = "inout",
                bottom = False)
plt.xlabel("Month", labelpad = 10) # tick label和axes的距离
plt.ylabel("Activation", labelpad = 10)
plt.plot(df["Month"],df["Activation"])
```

# 11. 数据补充表
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.table.html

* `matplotlib.pyplot.table(cellText=None, cellColours=None, cellLoc='right', colWidths=None, rowLabels=None, rowColours=None, rowLoc='left', colLabels=None, colColours=None, colLoc='center', loc='bottom', bbox=None, edges='closed', **kwargs) `

```python
cellText = [[1,2,3,4],[5,6,7,8]]# cellText2D list of str, optional
rows = ["row1","row2"] # rowLabelslist of str, optional
columns = ["col1","col2","col3","col4"] # rowLabelslist of str, optional
```
