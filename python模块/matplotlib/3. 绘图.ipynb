# 1. 折线图
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含两个子图的 Figure 对象
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# 在第一个子图中绘制 sin(x) 曲线
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
axes[0].plot(x, y, color='blue', linewidth=2)
axes[0].set_title('Sin(x)')

# 在第二个子图中绘制 cos(x) 曲线和直方图
y = np.cos(x)
axes[1].plot(x, y, color='red', linewidth=2)
axes[1].set_title('Cos(x)')

data = np.random.randn(1000)
bins = np.arange(-5, 5, 0.5)
axes[1].hist(data, bins=bins, color='gray')

# 设置全局标题
fig.suptitle('Sin(x) and Cos(x)')

# 显示图形
plt.show()
```

在使用 pyplot 绘图时，可以不需要显式地创建对象。这是因为 pyplot 库使用了内部状态机的设计模式，在调用绘图函数时会自动创建 Figure 和 Axes 对象，并且在每次调用绘图函数时，自动将绘图添加到当前 Axes 对象中。

例如，当我们使用 plt.plot(x, y) 绘制折线图时，pyplot 库会自动创建一个 Figure 对象和一个 Axes 对象，并将折线图绘制在当前的 Axes 对象中。如果在同一个 pyplot 会话中继续调用绘图函数，比如使用 plt.scatter(x, y) 绘制散点图，pyplot 库会自动将散点图添加到同一个 Axes 对象中。

虽然 pyplot 库的这种设计模式可以使绘图代码更加简洁，但是如果绘图操作比较复杂，或者需要更加精细的控制绘图的各个组件，建议使用更加面向对象的方式，显式地创建 Figure 和 Axes 对象，然后在 Axes 对象上调用各种绘图方法。这样可以更加清晰地组织代码，也更容易实现高级的绘图效果。

```python
matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

# linestyle or ls： {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
# marker：marker style string, Path or MarkerStyle
# line_width: linewidth or lw: float
# markeredgecolor or mec: color
# markerfacecolor or mfc: color #
# label = 图名字
```

# 面向对象显式绘图

```python


import matplotlib as mpl 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import random

mpl.pyplot.rcParams["font.sans-serif"] = "SimHei" # 解决中文乱码
mpl.pyplot.rcParams["axes.unicode_minus"] = False # 解决负号显示问题

df = pd.DataFrame({"Month":[1,2,3,4,5,6,7,8,9], 
                   "Registration":[866,2335,5170,6482,6120,1605,3813,4428,4631],
                   "Activation":[433,1167,2855,3241,3060,802,1902,2214,2315]})

# 1. 建立一个figure画图，和一个坐标系
fig, ax = plt.subplots()

# 2. 指明绘制数据
x = df["Month"]
y1 = df["Registration"]
y2 = df["Activation"]

# 3. 作图
ax.plot(x,y1,
        linestyle = "dashdot",
        linewidth = 1,
        marker = "o",
        markersize = 5,
        label = "the number of Registration")

# 4. 设置标题 and location
ax.set_title("XXX Company, 1-9 Month, Registration", loc="center")

# 5. 设置网格
ax.grid(visible=True, axis="x")

# 6. 为数据添加标签
for a,b in zip(x,y1):
    ax.text(a,b,b, ha="center", va="bottom", fontsize=10)

# 7. Legend
ax.legend()

# 保存
# plt.savefig("lu jing")
```

plt.title() 这个函数是 pyplot 模块中的一个函数，它可以直接被调用来为当前的图像添加一个标题。这个函数被设计成可以自动找到当前正在使用的 Axes 对象并对其添加标题。

在使用 pyplot 绘图的过程中，如果没有明确地创建一个 Axes 对象，pyplot 会自动创建一个默认的 Axes 对象，这个默认的 Axes 对象就是我们绘图时所看到的图形区域。

因此，当我们调用 plt.title() 时，其实是在自动找到当前正在使用的 Axes 对象并给它添加标题。这种方式非常方便，但在需要绘制多个子图的情况下，就需要通过显式创建 Axes 对象来更精确地控制绘图。

总的来说，当我们需要更细致地控制绘图时，使用显式的 Axes 对象是更好的选择，而当我们只需要简单地添加一个标题时，使用 plt.title() 更加方便。

axes 和 axis 是 matplotlib 中的两个不同的概念。

axes 是一个独立的绘图区域，它可以包含一个或多个图形元素（如线条、点等）以及其他绘图元素（如标题、标签等）。可以通过 fig.add_axes() 或 fig.subplots() 方法在 figure 中创建一个 axes 对象，也可以在一个已存在的 axes 上添加新的图形元素。

axis 表示绘图区域中的坐标轴。在 matplotlib 中，每个 axes 都有两个或三个 axis 对象，分别对应于水平轴（x轴）和垂直轴（y轴）。如果绘制的是三维图形，则还会有一个深度轴（z轴）。axis 包含着坐标轴上的刻度、标签等信息。

简而言之，axes 是整个绘图区域，而 axis 是 axes 上的坐标轴。

# 2. 柱状图
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
plt.subplot(1,1,1)
x = np.array(["East","West","North","South"])
y = np.array([8556,3125,7354,2534])

plt.bar(x,y, # The x coordinates of the bars. See also align for the alignment of the bars to the coordinates.
        # 如果传入的是个 array-like
       width = 0.5,
       align = "center", # 对齐
       label = "assignment"
       )

plt.title("Assignment in different place", loc = "center")

for a,b in zip(x,y):
    plt.text(a,b,b,
            ha = "center",
            va = "bottom",
            fontsize = 12)

plt.xlabel("Regin")
plt.ylabel("Assignment")
plt.legend()
```

在 matplotlib 中，bar() 函数的参数 x 可以是数字类型的数组或者是字符串类型的数组，因为 bar() 函数的作用就是在 x 坐标上绘制柱形图，x 轴可以是数字或者字符串类型。

如果 x 是数字类型的数组，则每个数字对应一个柱形图的位置；如果 x 是字符串类型的数组，则每个字符串对应一个柱形图的位置。在绘制柱形图时，需要按照 x 的顺序依次绘制柱形图，这样就可以将柱形图与其对应的 x 轴标签对应起来。如果不按顺序绘制柱形图，就会出现柱形图和标签不对应的情况，导致图形难以理解。

绘制柱形图时，还需要指定柱形的宽度、颜色、标签等参数。在 bar() 函数中，通过设置 width 参数可以指定柱形的宽度，通过设置 color 参数可以指定柱形的颜色，通过设置 label 参数可以为柱形添加标签。

# 3. 簇状图
在 Matplotlib 中，pyplot.bar() 方法绘制的柱状图可以接受一个包含字符串的数组作为横坐标，因为 Matplotlib 内部会将字符串转换成数字，并按照一定的比例绘制到图表上。

在 pyplot.bar() 方法中，参数 x 表示每个柱子所对应的横坐标位置。这里的横坐标位置并不是字符串，而是整数值。Matplotlib 内部会将传入的字符串数组转换成整数值，并根据 width 参数指定的柱子宽度绘制柱状图。因此，在传入字符串数组时，使用 x + 0.3 的方式仍然可以使得每个柱子之间保持一定的间隔，从而让图表更加美观易读。

```python
plt.subplot(1,1,1)
x = np.array([1,2,3,4])
y = np.array([8556,3125,7354,2534])
y1 = np.array([3124,2313,5432,1234])

# 1. 作图
plt.bar(x,y,
       width = 0.3,
       label = "assignment"
       )
plt.bar(x+0.3,y1,
        width = 0.3,
        label = "achievement"
       )

plt.title("Assignment&Achievement in different place", loc = "center")


# 添加数据标签
for a,b in zip(x,y):
    plt.text(a,b,b,
            ha = "center",
            va = "bottom",
            fontsize = 12)
    
for a,b in zip(x+0.3,y1):
    plt.text(a,b,b,
            ha = "center",
            va = "bottom",
            fontsize = 12)


plt.xlabel("Regin")
plt.ylabel("Assignment&Achievement")


plt.xticks(x+0.15, ["East","West","North","South"])

plt.legend()
```

If not None, autopct is a string or function used to label the wedges with their numeric value. The label will be placed inside the wedge. If autopct is a format string, the label will be fmt % pct. If autopct is a function, then it will be called.

```python
plt.subplot(1,1,1)
x = np.array([4123,4235,7546,2786])
labels = ["East","West","North","South"]

explode = [0.1,0,0,0] 
# If not None, is a len(x) array which specifies the fraction of the radius with which to offset each wedge.
# 里圆心句
labeldistance = 1.1
# The relative distance along the radius at which the labels are drawn. To draw the labels inside the pie, set labeldistance < 1. 

plt.title("?",loc = "center")
plt.pie(x,labels = labels, explode = explode, labeldistance = labeldistance,
       shadow = True,
        autopct = "%.2f%%",
        radius = 1
       )
plt.show()
```

在某些交互式环境中（例如 Jupyter Notebook 或 IPython Shell），Matplotlib 可以自动显示图形，而无需调用pyplot.show()函数。

这是因为这些环境通常具有 Matplotlib 的后端（backend），可以直接将图形显示在环境中。当你使用这些环境时，Matplotlib 会自动检测当前的后端，并根据需要自动显示图形。

然而，在其他环境中，例如 Python 脚本或标准命令行终端，Matplotlib 默认的后端是“无头”的（headless），不能自动显示图形。在这种情况下，你需要调用pyplot.show()函数才能看到绘制的图形。

因此，为了确保你的代码在所有环境中都能正常工作，最好在绘制图形后调用pyplot.show()函数，这样可以确保图形能够正常显示。

```python
plt.subplot(1,1,1, polar = True)

# labels
labels = ["python","excel","sql","tableau","skills"]

plt.title("SKILLS of a data analysist")

plt.polar(angles, # 角度
          data, # 不同角度的半径大小
          color ="r", marker = "o")

plt.xticks(ticks = angles[0:5], labels = labels)
# ticks
# array-like, optional
# The list of xtick locations. Passing an empty list removes all xticks.
#plt.thetagrids(np.arange(0, 360, 360/dataLenth), labels)

plt.show()
```

# 14. 热力图 correlation

```python
import itertools
cor_map = np.array([[1,0.5,0.3,0.1],
             [0.5,1,0.4,-0.1],
             [0.3,0.4,1,0.7],
             [0.1,-0.1,0.7,1]]) # 关系矩阵
color_cor_map = plt.cm.cool # 就是用来设置颜色地图的
plt.imshow(cor_map,cmap = color_cor_map)
plt.colorbar()

# ticks
tick_labels = ["data1","data2","data3","data4"]

ticks = np.arange(len(tick_labels))
plt.xticks(ticks = ticks,labels = tick_labels)
plt.yticks(ticks = ticks,labels = tick_labels)

# text np.array
for a,b in itertools.product(range(cor_map.shape[0]),
                             range(cor_map.shape[1])):
    plt.text(a,b,cor_map[a,b],ha = "center")
plt.grid(True) 
plt.show()
```
