## 3.1 建立画布 然后 画上坐标系

1. `subplot`: return a subplot axes positioned by the given grid definition
2. `subplot2grid`: create a subplot in a grid
3. `subplot_tool`: Launch a subplot tool window for a figure.
4. `subplots`: create a figure and a set of subplots

* figure可以理解成一块画布，subplot是众多小坐标系。
* `add_subplot`需要提前建立一个figure - 创建多个subplot并在此调用
* `subplot`和`subplot2grid`是一次将figure分成好几个subplot - 然后分别调用不同的subplot（每次调用只显示当先的subplot）
* `subplots`是创建一个figure和一系列subplot（全都显示出来）

* 第一种创建坐标系的方法`add_subplot`属于对象编程，所有操作针对某个对象进行的
* eg：先创建一块画布，然后在这块画布上去建立坐标系 进而在坐标系上进行绘图

```python
# 1. 建立画布

fig = pyplot.figure(figsize = (8,6)) # fig size 设置大小
#%matplotlib inline
# 返回内存地址
```

```python
# 2. 子画布
%matplotlib inline
fig = pyplot.figure()# 建立画布
ax1 = fig.add_subplot(2,2,1) # 画布上子画布：2行2列 第一个
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

## 3.2 直接建立坐标系（自动建立画布）
### 直接建立坐标系在画布上 `pyplot.subplot2grid( )`

```python
pyplot.subplot2grid((2,2),loc = (0,0)) # 建立一个两行两列子画图，显示第0行，第0列的坐标系

# 是一次将figure分成好几个subplot - 然后分别调用不同的subplot（每次调用只显示当先的subplot）
```

```python
import numpy as np
x = np.arange(6)
y = np.arange(6)
type(x)
```

```python
# 做一个2*2的坐标系，设置折线图
pyplot.subplot2grid((2,2),loc = (0,0))
# 是一次将figure分成好几个subplot - 然后分别调用不同的subplot（每次调用只显示当先的subplot）
pyplot.plot(x,y)
```

```python
# 做一个2*2的画布，设置折线图
pyplot.subplot2grid((2,2),loc = (0,1))
pyplot.bar(x,y)
```

```python
pyplot.subplot2grid((2,2),loc = (0,0))
pyplot.plot(x,y)
pyplot.subplot2grid((2,2),loc = (1,1))
pyplot.bar(x,y)
```

### `pyplot.subplot` 建立坐标系, show one

```python
from matplotlib import pyplot as plt
```

```python
plt.subplot(2,2,1) # 2*2 第一个坐标系
plt.plot(x,y)
# 是一次将figure分成好几个subplot - 然后分别调用不同的subplot（每次调用只显示当先的subplot）
plt.subplot(2,2,4)
plt.bar(x,y)
```

### 展示全部`plt.subplots(2,2)`

```python
plt.subplots(2,2) #是创建一个figure和一系列subplot（全都显示出来）
```

```python
axes = plt.subplots(2,2)
axes # 本质是个tuple
```

```python
fig, axes = plt.subplots(2,2) # 逗号
axes[0,1].plot(x,y)
axes[1,1].bar(x,y)
```
