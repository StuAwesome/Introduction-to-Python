# 3. 排名 - 根据现有列 rank
* `rank(ascending, method)`
* method: average first min max

```python
data = {"成交量": [10,20,10,30]}
df = pd.DataFrame(data)
df
```

```python
# 1. 默认平均值排名
df["成交量"].rank(ascending = False)
# 降序，1，2，3，4
# 但是第三名和第四名都是10，所以并列 = （3+4）/2
```

```python
df["成交量"].rank()
# 降序，1，2，3，4
# 但是第三名和第四名都是10，所以并列 = （3+4）/2
```

```python
# 2. first 关心谁先出现的
df["成交量"].rank(method = "first")

# index = 2，却是后出现的，所以rank = 2
```

```python
# 3. min 关心最小的
df["成交量"].rank(method = "min")
# 重复时都取小
```

```python
# 4. max 关心最大的
df["成交量"].rank(method = "max")
# 重复时都取大
```

# 12. 长宽表转换

```python
columns = ["Company","Name","Sale2013","Sale2014","Sale2015","Sale2016"]
data = [["apple","苹果",5000,5050,5050,5100],
        ["google","谷歌",3500,3800,3800,3900],
        ["facebook","脸书",2300,2900,2900,2900]
       ]
df = pd.DataFrame(data, columns = columns)
df
```

```python
# 设置下层次化索引
df = df.set_index(["Company","Name"])
df
```

```python
# 树形结构，先行再列
df = df.stack()
df
```

```python
# 更新索引
df.reset_index()
```

### 方法二 melt方法
1. `id_vars` 保持不变的列
2. `var_name` 新列的名字
3. `value_name` 新列值的名字

```python
columns = ["Company","Name","Sale2013","Sale2014","Sale2015","Sale2016"]
data = [["apple","苹果",5000,5050,5050,5100],
        ["google","谷歌",3500,3800,3800,3900],
        ["facebook","脸书",2300,2900,2900,2900]
       ]
df = pd.DataFrame(data, columns = columns)
df
```

```python
df.melt(id_vars = ["Company","Name"], var_name = "year", value_name = "sale")
```
