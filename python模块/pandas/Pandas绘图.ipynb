* pandas绘图很简单
* 基于data

```python
import pandas as pd
data = pd.DataFrame({"Month":[1,2,3,4,5], "Registration":[866,234,534,123,524],"Activation":[413,412,524,123,324]})
```

```python
data.set_index("Month", inplace = True)
```

```python
data
```

```python
help(data.plot)
# Parameter 很重要 很多都用的上！
```

```python
data.plot(kind = "line")
'''
- 'line' : line plot (default)
 |          - 'bar' : vertical bar plot
 |          - 'barh' : horizontal bar plot
 |          - 'hist' : histogram
 |          - 'box' : boxplot
 |          - 'kde' : Kernel Density Estimation plot
 |          - 'density' : same as 'kde'
 |          - 'area' : area plot
 |          - 'pie' : pie plot
 |          - 'scatter' : scatter plot (DataFrame only)
 |          - 'hexbin' : hexbin plot (DataFrame only)
 |      ax : matplotlib axes object, default None
'''
```

```python
'''
  subplots : bool, default False
 |          Make separate subplots for each column.
 '''


data.plot.line()
```

```python
data.plot.bar()
```

# 1. 折线图

```python
data.plot.line()
```

```python
data.plot.line(style = ["g","b"], fontsize = 15)
```

```python
data["Registration"].plot.line(style = ["g","b"], fontsize = 15,
                              use_index = False ,
                              grid = True)# 是否使用行索引做x轴
```

```python
data.plot.line(style = ["g","b"], 
               fontsize = 15,
               use_index = False,
               grid = True,
               subplots = True,
               layout = (2,1)
                              )# 是否使用行索引做x轴
```

```python
data.plot.line(style = ["g","b"], fontsize = 15,
              marker = "o")
```

# 2. 柱状图

```python
data.plot.bar()
```

```python
data.plot.barh()
```

# 3. 堆积柱状图

```python
data.plot.bar(stacked = True)
```

```python
data.plot.barh(stacked = True)
```

# 4. 直方图

```python
import numpy as np
df = pd.DataFrame({"a":np.random.randn(1000)+1, "b":np.random.randn(1000)
                  ,"c":np.random.randn(1000)-1}, columns = ["a","b","c"])
```

```python
df.plot.hist(stacked = True, bins = 20)
```

# 6. 面积图

```python
data.plot.area()
```

# 7. 散点图

```python
data.plot.scatter(x = "Registration", y= "Activation", color = "red")
```
